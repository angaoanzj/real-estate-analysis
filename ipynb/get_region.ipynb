```python
import pandas as pd
raw = pd.read_csv("code.csv")
```

```python
raw.count()
```

```python
#Delete duplicate place names that end with south east north west
filter_south = raw['place_name'].str.endswith('South')
raw1 = raw[~filter_south]
```

```python
raw1.count()
```

```python
filter_west = raw['place_name'].str.endswith('West')
raw[filter_west]
```

```python
raw1 = raw1[~filter_west]
raw1.count()
```

```python
filter_north = raw['place_name'].str.endswith('North')
raw1 = raw1[~filter_north]
raw1.count()
```

```python
filter_east = raw['place_name'].str.endswith('East')
raw1 = raw1[~filter_east]
```

```python
melb_code = []
```

```python
#connect code and city number
melb_code = raw1["place_name"].str.lower().str.split().str.join('-')+'-vic-'+raw1["postcode"].map(str)
```

```python
melb_code.to_csv('melb_code.csv')
```
